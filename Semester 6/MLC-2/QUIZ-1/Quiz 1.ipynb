```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X = np.random.rand(8,2)
X
```

```python
plt.scatter(X[:,0], X[:,1])
plt.title("Dataset Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
inertia = []

for k in range(1,7):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1, len(inertia)+1), inertia, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
import math

#Step
def step(x):
    if x >= 0:
        return 1
    else:
        return 0

#ReLU
def relu(x):
    return max(0, x)


#Hyperbolic Tangent
def tanh(x):
    return math.tanh(x)


# Input values
inputs = [-2, -1, 0, 1, 2]

print("Input Values:", inputs)
print()

# Apply activation functions
for x in inputs:
    print("Input =", x)
    print("Step Output :", step(x))
    print("ReLU Output :", relu(x))
    print("Tanh Output :", tanh(x))
    print()
```
